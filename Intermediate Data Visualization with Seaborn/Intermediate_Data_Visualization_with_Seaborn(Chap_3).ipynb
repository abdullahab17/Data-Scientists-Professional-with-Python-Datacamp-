```python
df = pd.read_csv('2010 US School Improvement Grants.csv')
print(df.head())
```

```python
# Create the stripplot
sns.stripplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         jitter=True)

plt.show()
```

```python
# Create and display a swarmplot with hue set to the Region
sns.swarmplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         hue='Region')

plt.show()
```

```python
# Create a boxplot
sns.boxplot(data=df,
         x='Award_Amount',
         y='Model Selected')

plt.show()
plt.clf()
```

```python
# Create a violinplot with the husl palette
sns.violinplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='husl')

plt.show()
plt.clf()
```

```python
# Create a boxenplot with the Paired palette and the Region column as the hue
sns.boxenplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='Paired',
         hue='Region')

plt.show()
plt.clf()
```

```python
# Show a countplot with the number of models used with each region a different color
sns.countplot(data=df,
         y="Model Selected",
         hue="Region")

plt.show()
plt.clf()
```

```python
# Create a pointplot and include the capsize in order to show caps on the error bars
sns.pointplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         capsize=.1)

plt.show()
plt.clf()
```

```python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         hue='Region')

plt.show()
plt.clf()
```

```python
tuition_df = pd.read_csv('2018 College Scorecard Tutuion.csv')
print(tuition_df.head())
```

```python
# Display a regression plot for Tuition
sns.regplot(data=tuition_df,
         y='Tuition',
         x='SAT_AVG_ALL',
         marker='^',
         color='g')

plt.show()
plt.clf()
```

```python
# Display the residual plot
sns.residplot(data=tuition_df,
          y='Tuition',
          x='SAT_AVG_ALL',
          color='g')

plt.show()
plt.clf()
```

```python
# The final plot should include a line using a 2nd order polynomial
sns.regplot(data=tuition_df,
            y='Tuition',
            x='PCTPELL',
            x_bins=5,
            order=2)

plt.show()
plt.clf()
```

```python
# Create a regplot and bin the data into 8 bins
sns.regplot(data=tuition_df,
         y='Tuition',
         x='UG',
         x_bins=8)

plt.show()
plt.clf()
```

```python
guest_df = pd.read_csv('Daily Show Guests.csv')
print(guest_df.head())
```

```python
# Create a crosstab table of the data
pd_crosstab = pd.crosstab(guest_df["Group"], guest_df["YEAR"])
print(pd_crosstab)

# Plot a heatmap of the table
sns.heatmap(pd_crosstab)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

plt.show()
```

```python
# Create the crosstab DataFrame
pd_crosstab = pd.crosstab(guest_df["Group"], guest_df["YEAR"])

# Plot a heatmap of the table with no color bar and using the BuGn palette
sns.heatmap(pd_crosstab, cbar=False, cmap="BuGn", linewidths=0.3)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

#Show the plot
plt.show()
plt.clf()
```
