```python
df = pd.read_csv('2010 US School Improvement Grants.csv')
print(df.head())
```

```python
df['Award_Amount'].plot.hist()
plt.show()
plt.clf()
```

```python
# Display a Seaborn displot
sns.displot(df['Award_Amount'])
plt.show()

# Clear the displot
plt.clf()
```

```python
# Create a displot
sns.displot(df['Award_Amount'],
             bins=20)

# Display the plot
plt.show()
```

```python
# Create a displot of the Award Amount
sns.displot(df['Award_Amount'],
             kind='kde',
             rug=True,
             fill=True)

# Plot the results
plt.show()
```

```python
insurance_df = pd.read_csv('Automobile Insurance Premiums.csv')
print(insurance_df.head())
```

```python
# Create a regression plot of premiums vs. insurance_losses
sns.regplot(data=insurance_df, x='insurance_losses', y='premiums')


# Display the plot
plt.show()
```

```python
# Create an lmplot of premiums vs. insurance_losses
sns.lmplot(data=insurance_df, x='insurance_losses', y='premiums')



# Display the second plot
plt.show()
```

```python
# Create a regression plot using hue
sns.lmplot(data=insurance_df,
           x="insurance_losses",
           y="premiums",
           hue="Region")

# Show the results
plt.show()
```

```python
# Create a regression plot with multiple rows
sns.lmplot(data=insurance_df,
           x="insurance_losses",
           y="premiums",
           row="Region")

# Show the plot
plt.show()
```
